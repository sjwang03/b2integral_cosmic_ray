# Prerequisites for ```binder``` execution

If you are runnning on ```binder``` you will be needing to install following python packages to excute data processing:

```gdown```

Therefore, be sure to execute the following cell every time you start a new ```binder``` session.

````python
import sys
import subprocess

# Run pip install quietly
subprocess.run(
    [sys.executable, "-m", "pip", "install", "gdown"],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL,
    check=True
)

print("Prerequisites installed successfully. Go on for data process.")
````

# Decompress using ```tar```

Here you need to specify the data file path and the target path where the decompressed data is put.

It is suggested that you use different cell when decompressing different data.

````python
import tarfile

tar_path = "../data/test/wfmdata.tar"

with tarfile.open(tar_path, "r:*") as tar:
    tar.extractall(path="../data/test/")
````
